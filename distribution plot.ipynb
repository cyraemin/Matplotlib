distribution plots are defined for quantitative variable which are continuous like age and salary etc.
displot/histplot
jointplot
pairplot
rugplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
#displot
sns.displot(df['total_bill'])
```

```python
#to change size of bins
sns.displot(df['total_bill'],bins=3)
```

```python
sns.displot(df['total_bill'],kde=True)
```

```python
plt.subplot(1,2,1)
sns.displot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.displot(df['tip'],bins=20,kde=True)
```

```python
#jointplot-to compare 2 data
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
#pairplot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')#to categorise data
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
#rugplot
sns.rugplot(df)
```

```python
sns.rugplot(df['total_bill'])
```
