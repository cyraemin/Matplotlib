```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

Matplotlib basics


```python
x=np.linspace(0,5,11)#linearly
x
y=x**3#cubed
y
```

```python
plt.plot(x,y)
plt.title("x v/s y")#graph name
plt.xlabel("value at x")#name for x axis
plt.ylabel("value at y")#name for y axis
```

subplots

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(x,y)#to plot in 4th plot
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x+2,y-3)



```

```python
plt.subplot(6,6,1)
plt.subplot(6,6,3)
plt.subplot(6,6,9)
plt.subplot(6,6,17)
plt.subplot(6,6,6)
```

object oriented way- to access the canvas and make changes in it
Matplotlib allows the aspect ratio, DPI, and figure size to be specified when the figure object is created. you can use the figsize and dpi keyword arguments.
figsize is a tuple of width and height of the figure in inches & dpi is dots  per inch(pixel per inch)


```python
fig=plt.figure(figsize=(10,8),dpi= 100)
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.1,0.1,0.8,0.8])
axis2.plot(x,y)
axis3=fig.add_axes([0.5,0.4,0.5,0.5])
axis3.plot(x,y)
```
